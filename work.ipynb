importing the dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

data collection and preprocessing

```python
#loading the data to a pd frame
raw_mail_data = pd.read_csv("C:/Users/hp/Documents/Projects/spam_mail_prediction/enron_spam_data.csv")
```

```python
print(raw_mail_data)
```

```python
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), ' ')
```

```python
print(mail_data.head())
print(raw_mail_data.shape)
```

label encoding
spam mail --> 0
ham mail --> 1

```python
mail_data.loc[mail_data['Spam/Ham'] == 'spam', 'Spam/Ham'] = 0
mail_data.loc[mail_data['Spam/Ham'] == 'ham', 'Spam/Ham'] = 1
```

```python
print(mail_data.head())
```

separating the data as text and labels

```python
x = mail_data['Message']

y = mail_data['Spam/Ham']
```

```python
print(y)
```

train test split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

feature  extraction

```python
# transform the text data to a feature vectors that can be used as input to logistic model
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

# convert y train and y test as intergers
y_trian = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features)
```

training the model

```python
model = LogisticRegression()
model.fit(x_train_features, y_trian)
```

evaluating the trained model

```python
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_dat = accuracy_score(y_trian, prediction_on_training_data)
```

```python
print(accuracy_on_training_dat)
```

```python
# prediction on test data
prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_dat = accuracy_score(y_test, prediction_on_test_data)
```

```python
print(accuracy_on_test_dat)
```
